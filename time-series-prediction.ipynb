```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("DATASET.xlsx")
df.head()
```

```python
df.columns = df.columns.str.replace('Unnamed: 0','Section')
df
```

```python
df = df.drop([910,911,912,913,914,915,916,917,918])
df.fillna(0,inplace=True)
df
```

```python
j=0
for i in range(100):
    df.iloc[j:j+10,0] = df.iloc[j,0]
    j+=10
df
```

```python
test = df.loc[df['year']==10]
test
```

```python
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import math
```

```python
etr = ExtraTreesRegressor(bootstrap=False, criterion='mse', max_depth=50,
                    max_features='auto', max_leaf_nodes=100,
                    min_impurity_decrease=0.0, 
                    min_samples_leaf=1,
                    min_weight_fraction_leaf=0.1,
                    n_jobs=None, oob_score=False, random_state=1, verbose=0,
                    warm_start=False)
```

```python
predicted = {'Para-9': [], 'Para-10': [], 'Para-11': [], 'Para-12': [], 'Para-13': []}
names = ['Para-9', 'Para-10', 'Para-11', 'Para-12', 'Para-13']
```

```python
rmse = []
N = 0
index = 0
for j in [10, 11, 12, 13, 14]:
  ans = 0
  count = 0
  for i in range(0, 1000, 10):
    xtrain = df.iloc[i:i+9, 1:10]
    ytrain = df.iloc[i:i+9, j]
    xtest = df.iloc[i+9, 1:10].to_numpy()
    ytest = df.iloc[i+9, j]
    etr.fit(xtrain, ytrain)
    ypred = etr.predict(xtest.reshape(1,-1))[0]
    predicted[names[index]].append(ypred)
    ans += (ytest - ypred)**2
    count += 1
    N += 1
  rmse.append(ans)
  index += 1
```

```python
df_pred = pd.DataFrame.from_dict(predicted)
df_pred
```

```python
for i in rmse:
  print((i/count)**0.5)
```

```python
print((sum(rmse)/N)**0.5)
```
